```python
from bertviz.attention_map import AttentionMapData, show
import torch
from mptb import BertModel, Config
from mptb.tokenization_juman import FullTokenizer
from IPython.display import display

tokenizer = FullTokenizer('data/Japanese_L-12_H-768_A-12_E-30_BPE/vocab.txt')
model_cfg = Config.from_json('config/bert_base.json', len(tokenizer), 512)
model = BertModel(model_cfg)
model.load_state_dict(torch.load('data/Japanese_L-12_H-768_A-12_E-30_BPE/Japanese_L-12_H-768_A-12_E-30_BPE.pt', map_location='cpu'))
```

```python
def call_html():
  import IPython
  display(IPython.core.display.HTML('''
        <script src="/static/components/requirejs/require.js"></script>
        <script>
          requirejs.config({
            paths: {
              base: '/static/base',
              "d3": "https://cdnjs.cloudflare.com/ajax/libs/d3/5.7.0/d3.min",
              jquery: '//ajax.googleapis.com/ajax/libs/jquery/3.3.1/jquery.min',
            },
          });
        </script>
        '''))
```

```python
sentence_a = "吾輩は猫である"
sentence_b = "名前はまだ無い"
map_data = AttentionMapData(model, tokenizer)
tokens_a, tokens_b, atts = map_data.get_data(sentence_a, sentence_b)
call_html()
show(tokens_a, tokens_b, atts)
```
